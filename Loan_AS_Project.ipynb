```python
#Checking and removing duplicate entries:

df.drop_duplicates(inplace=True)
df.shape
```

```python
numerics = ['int8', 'int16', 'int32', 'int64', 'float16', 'float32', 'float64']
categorical_columns=[]
features=df.columns.values.tolist()

for col in features:
    if df[col].dtypes in numerics:
        continue
    df[col] = LabelEncoder().fit_transform(df[col])
    
df
```

```python
#Multicollinearity using VIF

w = df[list(df.columns)]
vif_info=pd.DataFrame()
vif_info['VIF'] = [variance_inflation_factor(w.values,i) for i in range(w.shape[1])]
vif_info['Column'] = w.columns
vif_info.sort_values('VIF', ascending=False)
```

VIF Values are less which means we are good to proceed

##1st Quantile

```python
#1st Quantile
q1=df.quantile(0.25)

#3rd quantile
q3=df.quantile(0.75)

iqr=q3-q1
```

```python
ApplicantIncome_high=(q3.ApplicantIncome+(1.5*iqr.ApplicantIncome))
index=np.where(df['ApplicantIncome']>ApplicantIncome_high)
df=df.drop(df.index[index])
df.reset_index()
```

```python
CoapplicantIncome_high=(q3.CoapplicantIncome+(1.5*iqr.CoapplicantIncome))
index=np.where(df['CoapplicantIncome']>CoapplicantIncome_high)
df=df.drop(df.index[index])
df.reset_index()
```

```python
LoanAmount_high=(q3.LoanAmount+(1.5*iqr.LoanAmount))
index=np.where(df['LoanAmount']>LoanAmount_high)
df=df.drop(df.index[index])
df.reset_index()
```

```python
Loan_Amount_Term_high=(q3.Loan_Amount_Term+(1.5*iqr.Loan_Amount_Term))
index=np.where(df['Loan_Amount_Term']>Loan_Amount_Term_high)
df=df.drop(df.index[index])
df.reset_index()
```

#Balancing data using SMOTE

#SPLIT THE DATA

```python
x=df.drop('Loan_Status',axis=1)
y=df['Loan_Status']
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

#TRAIN_TEST_SPLIT

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=44)
```

```python
y_train.value_counts()
```

```python
sm=SMOTE()
```

```python
ove_smp=SMOTE(0.75)
x_train_ns,y_train_ns = ove_smp.fit_resample(x_train,y_train)
```

```python
y_train_ns.value_counts()
```

# MODEL BUILDING

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve,roc_auc_score
from sklearn.metrics import plot_roc_curve
```

```python
lr=LogisticRegression()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
```

```python
#Training of model

lr.fit(x_train_ns,y_train_ns)
knn.fit(x_train_ns,y_train_ns)
dt.fit(x_train_ns,y_train_ns)
rf.fit(x_train_ns,y_train_ns)

print('Training completed')
```

```python
##Scores checking

print(lr.score(x_test,y_test))
print(knn.score(x_test,y_test))
print(dt.score(x_test,y_test))
print(rf.score(x_test,y_test))
```

#The best model is Random Forest Model

```python
#Confusion Matrix

show=plot_roc_curve(dt,x_train,y_train)
plot_roc_curve(lr,x_train_ns,y_train_ns,ax=show.ax_)
plot_roc_curve(knn,x_train_ns,y_train_ns,ax=show.ax_)
plot_roc_curve(dt,x_train_ns,y_train_ns,ax=show.ax_)
plt.legend(prop={'size':10},loc='upper right')
plt.show()
```

```python


show=plot_roc_curve(dt,x_test,y_test)
plot_roc_curve(lr,x_test,y_test,ax=show.ax_)
plot_roc_curve(knn,x_test,y_test,ax=show.ax_)
plot_roc_curve(dt,x_test,y_test,ax=show.ax_)
plt.legend(prop={'size':10},loc='lower right')
plt.show()
```

```python
lr_pred=lr.predict(x_test)
knn_pred=lr.predict(x_test)
dt_pred=lr.predict(x_test)
rf_pred=lr.predict(x_test)
```

```python
print('Logistic Regression:',accuracy_score(y_test,lr_pred))
print('KNN:',accuracy_score(y_test,knn_pred))
print('Decision Tree:',accuracy_score(y_test,dt_pred))
print('Random Forest:',accuracy_score(y_test,rf_pred))
```

#RandomForest is the best model
